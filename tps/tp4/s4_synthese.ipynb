## Contexte et objectif

**Thème :** Immigration dans la presse belge (année 1920, *Journal de Bruxelles*).  

sujet de ce travail porte sur l’immigration dans la presse belge en 1920. J’ai choisi cette année car elle se situe juste après la Première Guerre mondiale : le climat socio‑politique est marqué par la reconstruction et les relations internationales, et je souhaitais voir comment l’immigration y est abordée. Ma question de recherche était de savoir quels termes, thématiques et tonalités apparaissent dans les articles qui parlent d’immigration en 1920. Je me suis demandé notamment si l’immigration était abordée d’un point de vue humanitaire (réfugiés, asile) ou plutôt administratif et économique (lois, naturalisation, finances). ## Méthode (rappel des étapes)

- **S1: Import du coprus**  J’ai d’abord importé la liste des fichiers .txt de 1920 contenus dans le dossier data/txt. Après vérification du chemin et de l’encodage, j’ai listé environ 100 fichiers correspondant à l’année 1920 et j’ai stocké la liste dans un fichier files_1920.json..

 - **S2 :Exploration/filtrage** :  J’ai filtré ces articles grâce à une liste de mots‑clés relatifs à l’immigration : immigration, immigrant(s), étranger/étrangère, réfugié(e), asile, naturalisation (j’ai ensuite ajouté visa et séjour pour élargir la recherche). Ce filtrage a permis d’identifier environ 80 articles « potentiellement liés à l’immigration » selon la présence de ces mots dans le texte. J’ai ensuite parcouru les titres et les premières lignes de quelques articles afin de choisir un article représentatif pour l’analyse fine. L’article retenu était un compte rendu parlementaire du 4 février 1920 qui traitait de la réduction des finances et des relations économiques avec l’étranger.
- **S3: Analyse** : 
  1.	Pré‑traitement et normalisation : j’ai chargé le texte de l’article choisi, passé le contenu en minuscules, supprimé les diacritiques (Un diacritique ou accent est un signe accompagnant une lettre ou un graphème pour en modifier le sens ou la prononciation.https://fr.wikipedia.org/wiki/Diacritique#:~:text=Un%20diacritique%2C%20signe%20diacritique%20(du%20grec%20ancien%C2%A0:%20%CE%B4%CE%B9%CE%B1%CE%BA%CF%81%CE%B9%CF%84%CE%B9%CE%BA%CF%8C%CF%82%C2%A0/%20diakritik%C3%B3s%2C%20qui%20distingue)%20ou%20accent%20est%20un%20signe%20accompagnant%20une%20lettre%20ou%20un%20graph%C3%A8me%20pour%20en%20modifier%20le%20sens%20ou%20la%20prononciation.) et extrait les tokens (mots) via une expression régulière. J’ai compté plus de 6 000 tokens bruts.
	2.	Nettoyage : j’ai supprimé les stopwords (mots vides comme le, la, ce, qui), les apostrophes et les tokens d’une seule lettre. J’ai conservé les tokens de longueur ≥ 2 pour l’analyse.
	3.	Concordances : pour chaque mot‑clé (immigration, immigré, étranger, réfugié, naturalisation, etc.), j’ai affiché des extraits du texte contenant le mot‑clé ainsi qu’un peu de contexte (15 mots à gauche et à droite). Cela m’a permis d’interpréter qualitativement le discours.
	4.	Collocations (bigrams) : j’ai calculé les paires de mots les plus fréquentes (bigrams) après nettoyage afin de repérer des expressions récurrentes.
	5.	Fréquences et visualisations : j’ai compté les fréquences des mots restants, affiché un bar‑chart des Top‑15 termes après stopwords, un autre bar‑chart après nettoyage approfondi, généré un nuage de mots et une série temporelle du nombre d’articles liés à l’immigration par mois (pour l’ensemble des articles filtrés de 1920). ## Résultats qualitatifs
- 1.	Nuage de mots (article choisi, après nettoyage)
Le nuage de mots offre une visualisation globale de la salience des termes. Les mots nous, que, pas, notre, belgique, guerre sont fortement mis en avant. Ces termes confirment que l’angle adopté est principalement institutionnel et national (référence à la Belgique, à la guerre et aux ministres) plutôt que centré sur les individus migrants. L’usage intensif des pronoms et connecteurs (nous, que, pas, dont, même) reflète la structure argumentative du texte et la perspective collective (« nous » belges).
- 2.	Concordances (en contexte)
Les extraits de texte pour chaque mot‑clé révèlent que le terme immigration apparaît surtout dans des phrases où l’on discute des relations économiques avec des pays étrangers ou de la réduction des dépenses publiques : « …les marchés étrangers constituent un danger pour nos finances publiques… », « …le nombre d’étrangers rendus par l’artifice d’un budget nous met dans les plus grands embarras… ». Le mot étranger est souvent accolé à des entreprises ou des paiements à l’étranger, illustrant l’enjeu du commerce. Le mot réfugié n’apparaît qu’une fois et concerne des dépenses pour des réfugiés russes. Le mot naturalisation figure dans une liste de naturalisés et n’est pas discuté en profondeur. Cela montre que l’article choisit de traiter l’immigration du point de vue des finances et de la régulation plutôt que des individus.
- 3.	Série temporelle des mentions (année 1920)
En compilant tous les articles filtrés par mots‑clés en 1920, j’ai compté le nombre d’articles liés à l’immigration par mois. Le graphique montre un pic notable en février et un autre en juin, correspondant possiblement à des débats parlementaires ou des événements diplomatiques. La courbe décroît ensuite et reste stable à partir de septembre. Cette tendance suggère que l’immigration n’est pas un sujet constant dans la presse de 1920 : elle réapparaît ponctuellement en lien avec l’actualité politique.

 ## Discussion et limites

Les résultats obtenus indiquent que l’immigration est abordée principalement sous un angle légal et administratif. Les listes de termes les plus fréquents, les collocations et les concordances montrent une préoccupation pour les ministres, les finances publiques et les relations économiques. On parle davantage du commerce extérieur et des paiements à l’étranger que des personnes migrantes elles‑mêmes. Cela laisse penser que l’immigration est perçue à travers les impacts macro‑économiques et géopolitiques plutôt qu’en termes humains.

Plusieurs limites méritent d’être soulignées. D’abord, le corpus analysé est relativement restreint : même si j’ai filtré environ 80 articles, je n’en ai étudié qu’un en profondeur. Le recours à l’OCR provoque des erreurs dans le texte qui peuvent fausser les analyses lexicales.  La recherche par mots‑clés reste simple et peut oublier des synonymes ou expressions proches (par exemple migrant, exil, expulsion). Le nettoyage des tokens élimine parfois des informations (par exemple, les formes conjuguées ou les noms propres). Enfin, l’analyse n’explore ni la tonalité (positive/négative) ni l’identification des acteurs (ministres, partis) de façon systématique.
