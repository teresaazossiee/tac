Je liste tous les fichiers .txt pour vérifier la structure et le volume.

```python
import os

corpus_path = "../data/txt/"
files = sorted(os.listdir(corpus_path))

print("Nombre total de fichiers :", len(files))
files[:10]
```

Je ne garde que les fichiers dont le nom contient “1920” (thématique ciblée).

```python
files_1920 = [f for f in files if "1920" in f]
print("Fichiers de 1920 :", len(files_1920))
files_1920[:10]
```

```python
with open(corpus_path + files_1920[0], "r") as f:
    print(f.read()[:1500])
```

```python
assert len(files) > 0 and len(files_1920) > 0, "Corpus ou sous-ensemble 1920 vide"
```

```python
import os, json
os.makedirs("./tp4/out", exist_ok=True)
with open("./tp4/out/files_1920.json", "w", encoding="utf-8") as f:
    json.dump(sorted(files_1920), f, ensure_ascii=False, indent=2)
print("Liste enregistrée dans ./tp4/out/files_1920.json")
```
